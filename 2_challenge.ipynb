#Produção de vinho por ano


```python

# Agrupe o DataFrame por ano e some a quantidade para cada ano
quantidade_total_por_ano = quantidade_por_ano_prod.groupby('Ano')['Qtd'].sum().reset_index()

# Configure o estilo do Seaborn
sns.set(style="whitegrid")

# Crie um gráfico de linha
plt.figure(figsize=(10, 6))
sns.lineplot(data=quantidade_total_por_ano, x='Ano', y='Qtd', marker='o', color='skyblue')

max_value = quantidade_total_por_ano['Qtd'].max()

# Defina o limite mínimo como zero e adicione um espaço acima
plt.ylim(0, max_value + 1000_000_0)  # Ajuste o valor 10000 conforme necessário

# Formate o eixo y para exibir números em milhões
def millions(x, pos):
    return '{:.0f}'.format(x / 1000_000)

formatter = FuncFormatter(millions)
plt.gca().yaxis.set_major_formatter(formatter)

# Adicione rótulos com os valores de cada ponto sem o "M"
for i in range(len(quantidade_total_por_ano)):
    plt.annotate(f'{quantidade_total_por_ano["Qtd"].iloc[i] / 1000000:.0f}',
                 (quantidade_total_por_ano["Ano"].iloc[i], quantidade_total_por_ano["Qtd"].iloc[i]),
                 textcoords="offset points", xytext=(0,10), ha='center')

# Configurações do gráfico
plt.title('Produção Anual de Vinho no Brasil')
plt.xlabel('Ano')
plt.ylabel('Quantidade (Milhões de litros) ')

# Exibe o gráfico de linha
plt.show()


```

#Variação anual de produção de vinho


```python
# Suponha que você tenha um DataFrame chamado 'producao_vinho' com as colunas 'Ano' e 'Qtd'
# Certifique-se de que os dados estejam em ordem cronológica

# Ordenar o DataFrame pelo ano
producao_vinho = quantidade_total_por_ano.sort_values(by='Ano')

# Calcular a variação anual em termos percentuais e adicionar o símbolo de porcentagem
producao_vinho['Variacao_Anuall'] = ((producao_vinho['Qtd'] - producao_vinho['Qtd'].shift(1)) / producao_vinho['Qtd'].shift(1) * 100).map('{:.2f}%'.format)

producao_vinho['Variacao_Anuall'] = producao_vinho['Variacao_Anuall'].str.replace('nan%', '-')

producao_vinho
```
